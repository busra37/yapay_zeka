```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df.dropna(subset=['Certificate','Meta_score','Gross'], inplace=True)
df.info()
```

```python
df.drop(['Poster_Link','Certificate','Runtime','Overview','Star1','Star2','Star3','Star4'],axis=1,inplace=True)
df.head()
```

```python
#gross (hasılat) sütunu object olarak gözüküyor. Bunu tamsayıya çevirelim.
df=df.convert_dtypes()
df['Gross']=df['Gross'].str.replace(',','')
df=df.astype({'Gross':'int'})
```
