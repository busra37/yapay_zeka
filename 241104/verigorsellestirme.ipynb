```python
import pandas as pd
data={
    "year":[2020,2021,2022,2023,2024],
    "sales (usd)":[50000,75000,10000,120000,150000],
    "customers":[120,180,250,300,400]
}
df=pd.DataFrame(data)
df
```

```python
plt.figure(figsize=(12,6))
```

```python
plt.ylabel("satıs usd")
plt.grid()
plt.subplot(1,2,2)
plt.bar(df["year"],df["customers"],color="orange")
plt.title("yıllık musteri sayısı")
plt.xlabel("yıl")
plt.ylabel("müsteri sayısı")
```

```python
plt.subplot(1,2,2)
plt.bar(df["year"],df["customers"],color="orange")
plt.title("yıllık musteri sayısı")
plt.xlabel("yıl")
plt.ylabel("müsteri sayısı")
```

```python
plt.pie(df["year"],labels=df["customers"])
plt.title("yıllık musteri sayısı")
plt.legend()
plt.show()
```
